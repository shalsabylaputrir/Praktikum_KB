## Mengatasi Nilai Null

```python
df=pd.read_csv("copydataset.csv")
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="most_frequent") 
```

```python
df["state"]=imputer.fit_transform(df[["state"]])
df["self_employed"]=imputer.fit_transform(df[["self_employed"]])
df["work_interfere"]=imputer.fit_transform(df[["work_interfere"]])
df["comments"]=imputer.fit_transform(df[["comments"]])
```
