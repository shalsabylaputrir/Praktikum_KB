```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("watson_healthcare_modified.csv")
df.head()
```

## Bar Plot

```python
sns.countplot(data=df,x = "Department")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x = 'Department',palette ='flare',hue='Education')
plt.title("Perbandingan Lamanya edukasi dengan setiap departement")
plt.grid()
plt.show()
```

data yang ditampilkan diatas dapat disimpulkan bahwa pada setiap departement rata-rata meraih pendidikan selama 3 tahun lamanya

### Grafik Headmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,cmap='Blues',annot_kws={'size':5})
plt.title('Korelasi antar pegawai kesehatan ')
plt.show()
```

Grafik HeatMap ini merupakan visualisasi atau pemetaan dengan menampilkan dari dataset Pegawai Kesehatan, yang menjelaskan korelasi antar variabel yang semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap.

### Grafik Pie

```python
df["Department"].value_counts()
```

```python
data_pie = df["Department"].value_counts().rename_axis("Department").reset_index(name="id_count")
```

```python
data_pie.head()
```

```python
plt.pie(data_pie["id_count"],labels=data_pie["Department"],startangle=90,autopct='%.1f%%',shadow=True,
        explode= (0.01,0,0)
       )
plt.show()
```

pada diagram grafik pie, menunjukkan bahwa pegawai terbanyak ada pada departement maternity dengan total 47.5% , diikuti dengan departement cardiology 31.7% dan neurology dengan total 20.8%
